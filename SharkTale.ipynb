#                                     The influence of the movie Shark Tale on shark attacks

Sharks always had a fierce and intimidating reputation until the movie Shark tale came out (01 Nov 2004), a movie that ridiculed sharks, specially White sharks, and made fun of them on the big screen. Did the (white) sharks feel like they had to prove themselves after this movie damaged their reputation?
Hypothesis: 
- Sharks in general attacked more people after the release of the movie to restore their reputation
- White sharks attacked more people after the release of the movie to restore their reputation
- Other species of sharks do not feel empathy for whtie sharks
- The movie influenced the way sharks attacked people



In this graphic we can see that the amount of shark attacks (by all species) in the 4 years before the release of the movie and the 4 years after the release are almost the same.




However the movie Shark Tale was specially making fun of White sharks, so let's take a look at the shark attacks done in the same period of time only by White sharks. Here we see there is an increase of attacks done by white sharks after the release of the movie. It seems like they did feel like they have to restore their reputation.



When we take a closer look at the attacks done in the 4 years before and 4 years after the release, we see that no other species had an increase even close to the one of the white sharks. So we can conclude that other shark species do not feel empathy for the white sharks and were not offended by the movie. 



Here we see the distrubition of attacks in the year before the release and year after. While before the movie there were months without reported white sharks attack, after the release (red line) the attacks have been more constant.



Shark attacks by all species do not follow this pattern. 



When we take a look at the attacks of white sharks, we see the biggest increase in numbers at attacks on surfers but the the attacks on fishers has been doubled.






When we look at the same statistics of all shark species, we see that all of them target fishers more. 

## Conclusion:



After the anaylsis we can conclude that White sharks felt offended by the way they were portraited in the movie Shark Tale. They had to do what they could do to get their fierce and killer reputation back; attack more people. About a year after the movie release it reached it's peak and they calmed down again. Other species of sharks did not feel empathy for the way white sharks image was damaged and they continued doing their own thing. There is no clear indication that the movie had influence on the activities people were doing during the attacks.
